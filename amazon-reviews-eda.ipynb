```python
data3.drop(['reviews.didPurchase','reviews.id','reviews.userCity','reviews.userProvince'], axis=1,inplace=True)
```

```python
#data3['column_name'].fillna(0, inplace=True)         # Text column
data3['name'].fillna('unknown',inplace=True)
```

```python
data3['asins'].fillna('B01AHB9CN2',inplace=True)
```

```python
data3.isnull().sum()
```

```python
data3['reviews.date'].fillna(method='ffill',inplace=True)
data3['reviews.dateAdded'].fillna(method='ffill',inplace=True)
data3['reviews.doRecommend'].fillna(data3['reviews.doRecommend'].mode()[0],inplace=True)
data3['reviews.numHelpful'].fillna(0,inplace=True)
data3['reviews.rating'].fillna(data3['reviews.rating'].median(),inplace=True)
data3['reviews.text'].fillna('no review text', inplace=True)
data3['reviews.title'].fillna('no title',inplace=True)
data3['reviews.username'].fillna('anonymous',inplace=True)
```

```python
data3.nunique()
```

```python
data3['reviews.date'] = pd.to_datetime(data3['reviews.date'], errors='coerce')
data3['review_year'] = data3['reviews.date'].dt.year
data3['review_month'] = data3['reviews.date'].dt.month
```

```python
data3
```

```python
data3 = data3[~data3['reviews.date'].isna()] #This keeps only rows where the date is valid
```

```python
data3.loc[:, 'review_length'] = data3['reviews.text'].astype(str).apply(len)
 #This line creates a new column called review_length, which stores the number of characters in each review.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(data3.columns)
```

```python
# Define sentiment labeling function
def label_sentiment(rating):
    if pd.isna(rating):
        return 'Unknown'
    elif rating >= 4:
        return 'Positive'
    elif rating == 3:
        return 'Neutral'
    else:
        return 'Negative'

# Apply function
data3 = data3.copy()
data3['sentiment'] = data3['reviews.rating'].apply(label_sentiment)
```

```python
sns.countplot(x='sentiment', data=data3, hue='sentiment', palette='Set2', legend=False)
plt.title('sentiment distribution')
plt.xlabel('sentiment')
plt.ylabel('number of reviews')
plt.show()
```

```python
data3['sentiment'] = data3['reviews.rating'].apply(lambda x: 'Positive' if x >= 4 else ('Neutral' if x == 3 else 'Negative'))
```

```python
data3.groupby('sentiment')['reviews.rating'].mean() # Average rating per sentiment 
```

```python
monthly_sentiment = data3.groupby(['review_year', 'review_month', 'sentiment']).size().unstack().fillna(0)
monthly_sentiment.plot(kind='bar', stacked=True, figsize=(12,6))
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='sentiment', hue='sentiment', data=data3, palette='Set2', legend=False)
plt.title('Distribution of Review Ratings')
plt.xlabel('Rating (1 = Worst, 5 = Best)')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

```python
monthly_reviews = data3.groupby(['review_year', 'review_month']).size().reset_index(name='review_count')
# Plot
plt.figure(figsize=(12,6))
sns.lineplot(data=monthly_reviews, x='review_month', y='review_count', hue='review_year', marker='o')
plt.title('Monthly Review Trend Over Years')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.xticks(range(1, 13))
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

```python
sentiment_trend = data3.groupby(['review_year', 'sentiment']).size().reset_index(name='count')

plt.figure(figsize=(10,6))
sns.lineplot(data=sentiment_trend, x='review_year', y='count', hue='sentiment', marker='o')
plt.title('Sentiment Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
sns.histplot(data3['review_length'], bins=50, kde=True, color='purple')
plt.title('Distribution of Review Length')
plt.xlabel('Review Length (in characters)')
plt.ylabel('Frequency')
plt.show()
```
